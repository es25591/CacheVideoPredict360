```python
class Environment(gym.Env):
    def __init__(self):
        super(Environment, self).__init__()
        
        self.action_space = spaces.Discrete(2)  # Example: two discrete actions
        self.observation_space = spaces.Box(
            low=0, 
            high=1, 
            shape=(4,), 
            dtype=np.float32
        )  # Example: 4D continuous observation
        
    def reset(self):
        self.state = np.random.rand(4)  # Example initial state
        return self.state
    
    def step(self, action):
        self.state = np.random.rand(4)
        reward     = np.random.rand()
        done       = np.random.rand() > 0.95        
        info       = {}
        
        return self.state, reward, done, info
    
    def render(self, mode='human'):
        """Render the environment to the screen."""
        pass  # Implement rendering logic if needed
```

```python
n = 100
env = Environment()

reset_res = env.reset()
obs = reset_res[0] if isinstance(reset_res, tuple) else reset_res

for step in range(n):
    action = env.action_space.sample()
    res = env.step(action)
    
    if len(res) == 4: # (obs, reward, done, info)
        obs, reward, done, info = res
    elif len(res) == 85: # (obs, reward, terminated, truncated, info)
        obs, reward, terminated, truncated, info = res
        done = terminated or truncated

    
    print(f"Step {step+1}/{n} - action: {action}, reward: {reward}, done: {done}")
    
    if done:
        break

print("Episode finished.")
```

```python
# Toy simulation: CPT-weighted vs Linear-weighted allocation for 360° tiles
# This code simulates tile importance, allocates a fixed bitrate budget proportionally
# to either linear weights (p*s) or CPT-weighted probabilities (w(p)*s), computes
# resulting tile quality, and compares perceived metrics using a CPT-style utility.
# It produces two plots (one per figure) and a summary DataFrame displayed to the user.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import log, exp

np.random.seed(42)

def prelec_weight(p, alpha=0.65):
    # Prelec one-parameter weighting: w(p)=exp(-(-ln p)^alpha)
    # Handle p==0 or 1 edge cases
    p = np.clip(p, 1e-12, 1-1e-12)
    return np.exp(-((-np.log(p))**alpha))

def cpt_utility(delta, loss_aversion=2.0, gain_power=0.5, loss_power=0.5):
    # delta = Qi - Q0
    # Gains: concave power, Losses: concave with loss aversion multiplier
    u = np.where(delta >= 0, (delta)**gain_power, -loss_aversion * ((-delta)**loss_power))
    return u

def single_run(N=100, B=1.0, b_full=0.02):
    # N tiles, total bitrate budget B, full-quality tile requires b_full bitrate
    # Generate attention probs and semantic saliency multiplier
    p = np.random.beta(2, 5, size=N)  # many low-prob tiles, some high
    s = np.random.uniform(0.6, 1.6, size=N)  # semantic saliency multipliers
    
    # Linear weights (naive): p * s
    w_lin = p * s
    # CPT-weighted probability
    w_p = prelec_weight(p, alpha=0.65)
    w_cpt = w_p * s
    
    # Allocate bitrate proportionally to weights
    b_lin = B * (w_lin / np.sum(w_lin))
    b_cpt = B * (w_cpt / np.sum(w_cpt))
    
    # Convert allocated bitrate to tile quality Qi in [0,1]
    Qi_lin = np.minimum(1.0, b_lin / b_full)
    Qi_cpt = np.minimum(1.0, b_cpt / b_full)
    
    # Reference expectation Q0 (use moving baseline: e.g., mean previous quality)
    # For toy simulation set Q0 to 0.5
    Q0 = 0.5
    
    # Compute CPT-utility per tile (human-perceived)
    u_lin = cpt_utility(Qi_lin - Q0)
    u_cpt = cpt_utility(Qi_cpt - Q0)
    
    # Perceived aggregate metric using true CPT probability weighting (what user "feels")
    p_weighted = prelec_weight(p, alpha=0.65)  # true subjective weighting
    perceived_lin = np.sum(p_weighted * u_lin)
    perceived_cpt = np.sum(p_weighted * u_cpt)
    
    # Also compute viewport-weighted quality (objective): sum p_i * Qi
    vwq_lin = np.sum(p * Qi_lin)
    vwq_cpt = np.sum(p * Qi_cpt)
    
    # Return relevant stats and arrays for optional inspection
    return {
        'perceived_lin': perceived_lin,
        'perceived_cpt': perceived_cpt,
        'vwq_lin': vwq_lin,
        'vwq_cpt': vwq_cpt,
        'p_mean': np.mean(p),
        'Qi_lin_mean': np.mean(Qi_lin),
        'Qi_cpt_mean': np.mean(Qi_cpt)
    }

# Run multiple trials
trials = 300
results = [single_run(N=100, B=1.0, b_full=0.02) for _ in range(trials)]
df = pd.DataFrame(results)

# Summary statistics
summary = df.agg(['mean', 'std']).T
summary = summary.rename(columns={'mean':'mean', 'std':'std'})

# Display a small table with mean ± std for key metrics
summary_table = pd.DataFrame({
    'metric': ['perceived (linear)', 'perceived (CPT)', 'VWQ (linear)', 'VWQ (CPT)'],
    'mean': [df['perceived_lin'].mean(), df['perceived_cpt'].mean(), df['vwq_lin'].mean(), df['vwq_cpt'].mean()],
    'std': [df['perceived_lin'].std(), df['perceived_cpt'].std(), df['vwq_lin'].std(), df['vwq_cpt'].std()]
})
summary_table['mean ± std'] = summary_table['mean'].round(4).astype(str) + ' ± ' + summary_table['std'].round(4).astype(str)

print("=" * 70)
print("CPT vs Linear Allocation Summary (toy simulation)")
print("=" * 70)
display(summary_table[['metric','mean ± std']])

# Plot 1: Bar chart with mean perceived metric and error bars
metrics = ['Perceived (linear)', 'Perceived (CPT)']
means = [df['perceived_lin'].mean(), df['perceived_cpt'].mean()]
errors = [df['perceived_lin'].std(), df['perceived_cpt'].std()]

plt.figure(figsize=(6,4))
plt.bar(metrics, means)
plt.errorbar(metrics, means, yerr=errors, fmt='none', capsize=8)
plt.ylabel('Aggregate perceived utility (CPT-weighted)')
plt.title('Perceived Utility: Linear vs CPT-based Allocation (mean ± std)')
plt.tight_layout()
plt.show()

# Plot 2: Bar chart for viewport-weighted quality (VWQ)
metrics2 = ['VWQ (linear)', 'VWQ (CPT)']
means2 = [df['vwq_lin'].mean(), df['vwq_cpt'].mean()]
errors2 = [df['vwq_lin'].std(), df['vwq_cpt'].std()]

plt.figure(figsize=(6,4))
plt.bar(metrics2, means2)
plt.errorbar(metrics2, means2, yerr=errors2, fmt='none', capsize=8)
plt.ylabel('Viewport-weighted quality (sum p_i * Qi)')
plt.title('Objective viewport-weighted quality: Linear vs CPT allocation (mean ± std)')
plt.tight_layout()
plt.show()

# Also provide the raw DataFrame
print("\n" + "=" * 70)
print("Raw trial results (first 30 rows)")
print("=" * 70)
display(df.head(30))

# Provide a brief textual summary for the user
print("\n" + "=" * 70)
print("Simulation Results Summary")
print("=" * 70)
print("Simulation complete: 300 trials. Table(s) and plots displayed. Interpretation guidance:")
print("- 'Perceived' metric is the aggregate CPT-weighted utility using the Prelec weighting as the user's subjective attention model and a loss-averse utility.")
print("- Higher 'Perceived' means the allocation better matches human subjective value (what the user 'feels').")
print("- 'VWQ' is an objective viewport-weighted quality; higher is better from an objective standpoint.")
print("\nObserved (toy simulation) results:")
print(f"Mean perceived (linear): {df['perceived_lin'].mean():.4f} ± {df['perceived_lin'].std():.4f}")
print(f"Mean perceived (CPT):    {df['perceived_cpt'].mean():.4f} ± {df['perceived_cpt'].std():.4f}")
print(f"Mean VWQ (linear):       {df['vwq_lin'].mean():.4f} ± {df['vwq_lin'].std():.4f}")
print(f"Mean VWQ (CPT):          {df['vwq_cpt'].mean():.4f} ± {df['vwq_cpt'].std():.4f}")
```

```python
# Simulation: CPT-driven tile prefetch selection for single user (toy example)
# - We enumerate all subsets (exact solution) for small number of tiles
# - Compare CPT-based selection vs expected-utility (EUT) selection (baseline)
# - Easy to follow, parameterizable

import itertools, math, random
import numpy as np
import pandas as pd

random.seed(1)
np.random.seed(1)

# Parameters
T = 12  # number of tiles (small for enumeration)
cache_capacity = 10  # total cache units available

# Tile sizes (units of cache cost)
sizes = np.random.randint(1, 4, size=T)  # sizes in {1,2,3}

# Predicted view probabilities p_t (from viewport+semantic predictor) - simulated
# Let's bias some tiles as semantically salient (higher p)
base_probs = np.clip(np.random.beta(2,5,size=T), 0.01, 0.99)
saliency = np.zeros(T)
saliency[[1,3,7]] = 0.6  # mark some tiles as semantically important
p = np.clip(base_probs + saliency, 0.01, 0.99)

# Outcome values when prefetching:
# - If the tile is viewed and was prefetched: benefit b (gain)
# - If the tile is NOT viewed but prefetched: wasted cost w (loss)
# These are QoE-delta units. You can tune per tile or keep constant.
b = np.full(T, 3.0)   # benefit when prefetched & viewed
w_cost = np.full(T, 0.8)  # wasted cost when prefetched & not viewed

# CPT parameters (typical values from literature; adjustable)
alpha = 0.88  # gain curvature
beta = 0.88   # loss curvature
lam = 2.25    # loss aversion (losses weighted more)
prelec_gamma = 0.65  # probability weighting parameter (Prelec)

# Define utility functions
def u_gain(x):
    # x must be >= 0
    return x**alpha

def u_loss(x):
    # x must be < 0; returns negative value magnitude times lambda
    return -lam * ((-x)**beta)

def prelec_w(p):
    # Prelec weighting function: w(p) = exp(-(-ln p)^gamma)
    # For p=0 or 1 handle edges
    p = np.clip(p, 1e-12, 1-1e-12)
    return math.exp(-((-math.log(p))**prelec_gamma))

# Compute CPT expected utility of prefetching each tile individually
cpt_value_tile = np.zeros(T)
eut_value_tile = np.zeros(T)

for t in range(T):
    pt = p[t]
    # outcomes: gain = +b[t] with prob pt; loss = -w_cost[t] with prob 1-pt
    gain = b[t]
    loss = -w_cost[t]
    # CPT: weighted sum of value function applied to gains/losses
    w_pt = prelec_w(pt)
    w_1pt = prelec_w(1-pt)
    cpt_value_tile[t] = w_pt * u_gain(gain) + w_1pt * u_loss(loss)
    # EUT baseline: expected (linear) utility (here simply expected net QoE)
    eut_value_tile[t] = pt * gain + (1-pt) * (-w_cost[t])

# Now solve the 0-1 knapsack by enumeration (exact) for both CPT objective and EUT
best_cpt = {"value": -1e9, "subset": None, "size": None}
best_eut = {"value": -1e9, "subset": None, "size": None}

for r in range(1<<T):  # all subsets
    chosen = [(r>>i)&1 for i in range(T)]
    total_size = sum(sizes[i] for i in range(T) if chosen[i])
    if total_size > cache_capacity:
        continue
    # objective values
    val_cpt = sum(cpt_value_tile[i] for i in range(T) if chosen[i])
    val_eut = sum(eut_value_tile[i] for i in range(T) if chosen[i])
    if val_cpt > best_cpt["value"]:
        best_cpt.update({"value": val_cpt, "subset": chosen.copy(), "size": total_size})
    if val_eut > best_eut["value"]:
        best_eut.update({"value": val_eut, "subset": chosen.copy(), "size": total_size})

# Prepare table for display
df = pd.DataFrame({
    "tile": np.arange(T),
    "size": sizes,
    "p(view)": np.round(p,3),
    "benefit_if_viewed": b,
    "wasted_cost_if_not_viewed": w_cost,
    "CPT_value_tile": np.round(cpt_value_tile,4),
    "EUT_value_tile": np.round(eut_value_tile,4)
})

# Which tiles selected by each method
cpt_selected = [i for i,x in enumerate(best_cpt["subset"]) if x]
eut_selected = [i for i,x in enumerate(best_eut["subset"]) if x]

# Display results
import caas_jupyter_tools as tools
tools.display_dataframe_to_user("Tile parameters and per-tile utilities", df)

print("Cache capacity:", cache_capacity)
print("\nBest CPT solution:")
print(" Selected tiles:", cpt_selected)
print(" Total cache used:", best_cpt["size"])
print(" Total CPT objective value:", round(best_cpt["value"],4))

print("\nBest EUT solution:")
print(" Selected tiles:", eut_selected)
print(" Total cache used:", best_eut["size"])
print(" Total EUT objective value:", round(best_eut["value"],4))

# Also show the sets side-by-side for clarity
def subset_str(subset):
    return "[" + ", ".join(str(i) for i,x in enumerate(subset) if x) + "]"

print("\nSubset bitmask (CPT):", subset_str(best_cpt["subset"]))
print("Subset bitmask (EUT):", subset_str(best_eut["subset"]))

# Save a CSV for later inspection
df.to_csv("/mnt/data/tile_parameters.csv", index=False)
print("\nSaved tile parameters to /mnt/data/tile_parameters.csv")
```

```python
# Retry: Implement a minimal Gym-like CPTPrefetchEnv WITHOUT importing gymnasium (for environments where gymnasium is not installed).
# This preserves the same API (reset, step, render) and runs the step-by-step demo.

from typing import Tuple, Dict
import numpy as np
import math
import pandas as pd

# -------------------- CPT functions --------------------
def prelec_weight(p: float, gamma: float = 0.7) -> float:
    p = np.clip(p, 1e-9, 1.0)
    return math.exp(-((-math.log(p)) ** gamma))

def cpt_value(x: float, alpha: float = 0.88, beta: float = 0.88, lam: float = 2.25) -> float:
    if x >= 0:
        return x ** alpha
    else:
        return -lam * ((-x) ** beta)

# -------------------- Minimal Gym-like environment --------------------
class CPTPrefetchEnvSimple:
    """
    Minimal Gym-like CPT prefetch environment (no external gym dependency).
    Observation: dict { 'p': np.array(T,), 'capacity': scalar, 'step': scalar }
    Action: binary array length T
    """
    def __init__(self, num_tiles: int = 8, capacity: int = 3, seed: int | None = 0, max_steps: int = 20):
        self.num_tiles = num_tiles
        self.capacity = int(capacity)
        self.rng = np.random.default_rng(seed)
        self.gain_if_prefetched = 1.0
        self.loss_if_not_prefetched = -2.0
        self.gamma = 0.7
        self.alpha = 0.88
        self.beta = 0.88
        self.lam = 2.25
        self.current_p = np.zeros(self.num_tiles, dtype=float)
        self.current_step = 0
        self.max_steps = max_steps

    def reset(self, seed: int | None = None):
        if seed is not None:
            self.rng = np.random.default_rng(seed)
        base = self.rng.uniform(0.01, 0.2, size=self.num_tiles)
        salient_indices = self.rng.choice(self.num_tiles, size=max(1, self.num_tiles//4), replace=False)
        base[salient_indices] += self.rng.uniform(0.4, 0.8, size=salient_indices.size)
        self.current_p = np.clip(base, 0.0, 1.0)
        self.current_step = 0
        obs = {"p": self.current_p.copy(), "capacity": self.capacity, "step": self.current_step}
        return obs

    def compute_expected_cpt_reward(self, action: np.ndarray) -> float:
        reward = 0.0
        for t in range(self.num_tiles):
            p = float(self.current_p[t])
            w = prelec_weight(p, gamma=self.gamma)
            if action[t] == 1:
                expected_outcome = p * self.gain_if_prefetched
            else:
                expected_outcome = p * self.loss_if_not_prefetched
            v = cpt_value(expected_outcome, alpha=self.alpha, beta=self.beta, lam=self.lam)
            reward += w * v
        return float(reward)

    def step(self, action: np.ndarray):
        action = np.array(action).astype(int).reshape(-1)[:self.num_tiles]
        n_prefetch = int(action.sum())
        info = {
            "n_prefetch_requested": n_prefetch, 
            "capacity": self.capacity, 
            "prefetch_indices": np.where(action==1)[0].tolist()
        }

        reward = 0.0
        done = False

        if n_prefetch > self.capacity:
            selected = np.where(action == 1)[0]
            to_drop = n_prefetch - self.capacity
            sel_sorted = selected[np.argsort(self.current_p[selected])]
            drop = sel_sorted[:to_drop]
            action[drop] = 0
            info["clipped_dropped_indices"] = drop.tolist()
            penalty = -0.5 * to_drop
            reward += penalty

        reward += self.compute_expected_cpt_reward(action)

        self.current_step += 1
        noise = self.rng.normal(0, 0.02, size=self.num_tiles)
        self.current_p = np.clip(self.current_p + noise, 0.0, 1.0)

        if self.current_step >= self.max_steps:
            done = True

        obs = {"p": self.current_p.copy(), "capacity": self.capacity, "step": self.current_step}
        return obs, reward, done, info

    def render(self):
        saliency = ", ".join([f"{p:.2f}" for p in self.current_p])
        print(f"[Step {self.current_step}] p_t = [{saliency}] (capacity={self.capacity})")

# -------------------- Demo run --------------------
def demo_run_simple(seed: int = 0):
    env = CPTPrefetchEnvSimple(num_tiles=8, capacity=3, seed=seed, max_steps=10)
    obs = env.reset(seed=seed)
    print("Initial observation (tile probabilities):")
    print(pd.DataFrame({
        "tile": np.arange(env.num_tiles),
        "p_t (predicted)": np.round(obs["p"], 3),
    }))

    print("\n--- Running episode step-by-step ---\n")
    done = False
    step = 0
    history = []
    while not done and step < 10:
        p = obs["p"]
        request_k = env.capacity + (1 if step % 3 == 0 else 0)
        topk_idx = np.argsort(-p)[:request_k]
        action = np.zeros(env.num_tiles, dtype=int)
        action[topk_idx] = 1

        obs, reward, done, info = env.step(action)
        env.render()

        print(f"Action requested (top-{request_k}): {topk_idx.tolist()}")
        if "clipped_dropped_indices" in info:
            print(" -> clipped because of capacity. dropped:", info["clipped_dropped_indices"])
        print(f"Received reward: {reward:.4f}  (n_prefetch_requested={info['n_prefetch_requested']})\n")

        history.append({
            "step": step,
            "requested": info["n_prefetch_requested"],
            "prefetched": env.capacity if info.get("clipped_dropped_indices") else info["n_prefetch_requested"],
            "reward": reward,
            "prefetch_indices": info["prefetch_indices"],
        })
        step += 1

    print("Episode history (summary):")
    df_hist = pd.DataFrame(history)
    print(df_hist.to_string(index=False))

demo_run_simple(seed=42)
```
